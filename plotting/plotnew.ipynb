```python
import csv
import glob
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
def getAve(arr):
    return sum([int(x) for x in arr]) / len(arr)

def getListOfFiles(re, _pre = 7, _post = 4):
    return sorted(glob.glob(re), key=lambda s: int(s[_pre:-1 * _post]))

def getFirstRowOfCsv(file_name):
    with open(file_name, newline='') as csvfile:
        spamreader = csv.reader(csvfile, delimiter=',')
        # Retun the first row
        for row in spamreader:
            return row
```

```python
### Global Variables
y = []
err = []
fileRE = '../experiments/Exp3-local-ganache-done/Exp3-local-*.csv'
preChar = 26 + 11 + 13
postChar = 4
```

```python
### Main Execution
for file_name in getListOfFiles(fileRE,preChar,postChar):
    row = np.asarray(getFirstRowOfCsv(file_name))
    
    
    y.append(getAve(row))
    err.append(np.std(row.astype(np.float)))
# #     ans.append(getAve(row))
#     print(file_name, y, err)
```

```python
x = np.arange(10,101,10)
y = np.asarray(y)
print(y)
print(x)
print(err)

plt.figure(num=None, figsize=(20, 12), dpi=80, facecolor='w', edgecolor='k')
plt.rcParams.update({'font.size': 18})
plt.xticks(np.arange(0, 110, step=10))
plt.yticks(np.arange(0, 2000, step=250))
plt.xlabel("Participation Threshold (%)")
plt.errorbar(x, y, err, fmt="ok", capsize=5, ecolor="grey");
```
